```python
#3. 감염 시뮬레이션 확장:2단계 이내 실습 (TC - 1)
from collections import deque

def find_infections_bfs(graph, start_node, max_level=2):
    """
    BFS를 사용하여 특정 노드로부터 max_level 이내의 모든 노드를 찾습니다.
    (감염 경로 추적)

    :param graph: 인접 리스트 형태의 그래프 (dict)
    :param start_node: 최초 감염자 (시작 노드)
    :param max_level: 찾고자 하는 최대 깊이(단계)
    :return: (감염자, 감염 레벨) 튜플의 리스트
    """

    # 1. 큐(Queue)와 방문(Visited) 기록, 결과 리스트 초기화
    # 큐에는 (사람, 현재 레벨) 튜플을 저장합니다.
    queue = deque([(start_node, 0)])

    # visited는 방문 여부와 함께 *최단* 레벨을 저장할 수 있습니다.
    # 시작 노드는 0레벨로 방문 처리
    visited = {start_node: 0}

    # 결과 리스트 (감염자 자신은 제외)
    infected_list = []

    # 2. 큐가 빌 때까지 BFS 탐색 시작
    while queue:
        current_person, current_level = queue.popleft()

        # 3. 최대 레벨(2단계)에 도달하면 더 이상 탐색하지 않음
        # (현재 노드의 *다음* 레벨이 max_level을 초과하면 안 됨)
        if current_level >= max_level:
            continue

        # 4. 현재 사람의 모든 접촉자(이웃) 확인
        # .get(current_person, [])는 키가 없는 노드(예: 접촉자가 없는 사람)일 경우
        # 오류 대신 빈 리스트를 반환하여 안전하게 처리합니다.
        for neighbor in graph.get(current_person, []):

            # 5. 아직 방문하지 않은 사람인 경우
            if neighbor not in visited:

                #Level 갱신
                new_level = current_level + 1

                # 방문 처리 (해당 노드의 레벨 기록)
                visited[neighbor] = new_level

                # 큐에 추가하여 다음 탐색 대상으로 등록
                queue.append((neighbor, new_level))

                # 결과 리스트에 (감염자, 레벨) 추가 (조건 2, 3 충족)
                infected_list.append((neighbor, new_level))

    return infected_list

# --- 제공된 그래프 데이터 ---
new_graph = {
    '고용민': ['김진서', '양지원', '강예근'],
    '김진서': ['박진영', '이유라'],
    '양지원': ['김영휘'],
    '김영휘': ['고용민', '강수진'],
    '강예근': ['이유라', '박민혁'],
    '이유라': ['이서경'],
    '이서경': ['박민혁'],
    '박민혁': ['강예근'],
    '박진영': ['김예진'],
    '김예진': ['양지원'],
    '강수진': ['박진영']
}

# --- 시뮬레이션 실행 ---
start_person = '고용민'
infections = find_infections_bfs(new_graph, start_person, max_level=2)

# --- 결과 출력 (형식 자유) ---
print(f"--- 💻 '{start_person}' 기준 2단계 이내 감염 경로 추적 (BFS) ---")
print(f"최초 감염자: {start_person}\n")

# 레벨별로 보기 좋게 그룹화
level_1 = [person for person, level in infections if level == 1]
level_2 = [person for person, level in infections if level == 2]

print(f"🔵 1단계 감염 (거리 1):")
if level_1:
    print(f"   {', '.join(level_1)}")
else:
    print("   (없음)")

print(f"\n🔴 2단계 감염 (거리 2):")
if level_2:
    print(f"   {', '.join(level_2)}")
else:
    print("   (없음)")

print("\n--- 상세 결과 (사람, 감염 단계) ---")
for person, level in infections:
    print(f"   {person} (감염 단계: {level})")
```
